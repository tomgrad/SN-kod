[Hindmarsh-Rose model (wiki)](https://en.wikipedia.org/wiki/Hindmarsh%E2%80%93Rose_model)

```python
%pylab inline
from scipy.integrate import odeint
```

```python
def hr(X,t):
    s, x0, a, b, c, d = 4, -1.6, 1, 3, 1, 5
    r=0.02 # 0.001-0.035
    I=3 # 2.3-3.42
    
    x,y,z=X
    dx = y-a*x**3+b*x**2-z+I
    dy = c-d*x**2-y
    dz = r*(s*(x-x0)-z)
    return dx, dy, dz
```

```python
T=500
dt=0.01
t = np.linspace(0, T, num=int(T/dt))
x0=array([-1.2, -7, 2.6])

x = odeint(hr, x0, t)
```

```python
subplots(figsize=(15, 5))
subplot(121)
plot(x[:,0],x[:,1])
subplot(122)
plot(t,x[:,0]);
```
