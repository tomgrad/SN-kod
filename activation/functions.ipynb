```python
x=linspace(-5, 5, 128)
```

```python
# logistic function

def logistic(x):
    return 1/(1+exp(-x))

plot(x, logistic(x))
grid()
# savefig("logistic.svg")
```

```python
#tanh
plot(x, tanh(x))
grid()
# savefig("tanh.svg")
```

```python
# rectified linear unit (relu)
plot(x, maximum(0, x))
grid()
# savefig("relu.svg")
```

```python
# leaky relu
plot(x, maximum(x/10, x))
grid()
# savefig("leaky_relu.svg")
```

```python
def elu(x, α):
    return where(x>0, x, α*(exp(x)-1))

plot(x, elu(x, 1))
grid()
# savefig("elu.svg")
```
